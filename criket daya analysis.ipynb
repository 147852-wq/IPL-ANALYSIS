**TOSS DECISION**

```python
teams = df["toss_winner"].unique()
teams
```

```python
teams = df["toss_winner"].unique()
decision_making = pd.DataFrame([], columns=["Toss Winner", "Decision", "Times"])
for id,t in enumerate(teams):
    temp_bat = df[(df["toss_winner"]== t) & ((df["toss_decision"]== "bat"))]
    temp_field = df[(df["toss_winner"]== t) & ((df["toss_decision"]== "field"))]
    decision_making = pd.concat([decision_making, pd.DataFrame([{"Toss Winner":t, "Decision":"bat", "Times":temp_bat["toss_winner"].count()}])], ignore_index=True)
    decision_making = pd.concat([decision_making, pd.DataFrame([{"Toss Winner":t, "Decision":"field", "Times":temp_field["toss_winner"].count()}])], ignore_index=True)
```

```python
decision_making
```

```python
sns.catplot(x= "Toss Winner", y="Times", hue="Decision", data=decision_making, kind="bar", height=8.27, aspect=20/8.27)
plt.xticks(rotation=90)
plt.title("Toss Decision by Team")
plt.show()
```

```python
sns.catplot(
    x="Toss Winner",
    y="Times",
    hue="Decision",
    data=decision_making,
    kind="bar",
    height=5,
    aspect=2
)
plt.xticks(rotation=90)
plt.xlabel("IPL Teams")
plt.ylabel("Toss Decision")
plt.title("IPL Toss Analysis")
plt.show()
```
